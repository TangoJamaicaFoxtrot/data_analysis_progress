```python
churned_customers = sales_df.loc[
(sales_df["Renewal_Status"]=="Churned"),
["Customer_ID", "Region", "Churn_Flag"]
]
display(churned_customers)

# okay, so looking at this dataset - churn flag is not corresponding with "Renewal Status"
# so next I'm going to merge the datasets and then edit to ensure that Renewed = 0 and Churned = 1
```

```python
merged_df = pd.merge(sales_df, customer_df, on="Customer_ID", how="inner")
merged_df.head()
```

```python
mismatch_df = merged_df[(merged_df["Churn_Flag"]==1) & (merged_df["Renewal_Status"] != "Churned")]
display(mismatch_df)
# okay we've identified a whole bunch of incorrect data - next cell we can use lambda to edit the Churn Flag data
```

```python
merged_df["Churn_Flag"] = merged_df["Renewal_Status"].apply(lambda x:1 if x=="Churned" else 0)
display(merged_df)
# Here we've replaced the incorrect data. Let's run the mismatch_df next to see if we've cleaned them all up
```

```python
mismatch_df = merged_df[(merged_df["Churn_Flag"]==1) & (merged_df["Renewal_Status"] != "Churned")]
display(mismatch_df)
# Boom! No cells exist where Renewed and 1 in Churn Flag. That's a data cleaning exercise for sure
```

```python
segment_plan_summary = merged_df.groupby(["Customer_Segment", "Subscription_Plan"]).agg({
    "Monthly_Revenue":"mean",
    "Product_Usage_Score":"mean"
}).rename(columns={
    "Monthly_Revenue":"Avg_Monthly_Revenue",
    "Product_Usage_Score":"Avg_Product_Usage_Score"
})
display(segment_plan_summary)
```

```python
segment_plan_summary["Customer_Count"] = merged_df.groupby(["Customer_Segment", "Subscription_Plan"]).agg({
    "Customer_ID":"size"
})
display(segment_plan_summary)
```

```python
segment_plan_summary.pivot_table(values="Customer_Count", index="Customer_Segment", columns="Subscription_Plan").plot(kind="bar", stacked=True, figsize=(10,6))
plt.title("Customer Count by Segment and Subscription Plan")
plt.xlabel("Customer Segment")
plt.ylabel("Number of Customers")
plt.xticks(rotation=0)
plt.legend(title="Subscription Plan", bbox_to_anchor=(1.19,1), loc="upper right")
```

```python
plt.figure(figsize=(10,6))
revenue_pivot = segment_plan_summary.pivot_table(values="Avg_Monthly_Revenue", index="Customer_Segment", columns="Subscription_Plan")
sns.heatmap(revenue_pivot, annot=True, cmap="Blues", fmt=".2f")
plt.title("Average Monthly Revenue by Customer Segment and Subscription Plan")
plt.xlabel("Subscription Plan")
plt.ylabel("Customer Segment")
plt.xticks(rotation=0)
plt.show()
```
