```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("/Users/anastasia/Downloads/erroneous_data.csv")
display(df.info())
```

```python
# convert columns to arrays for NumPy
revenue_array = df["Revenue"].values
print("Revenue Array:\n")
display(revenue_array)
```

```python
# filter out NaN & negatives
valid_revenue = revenue_array[~np.isnan(revenue_array) & (revenue_array >= 0)]
print("Filtered Revenue Array:\n\n", valid_revenue)
```

```python
# mean, median, std dev calc
mean_revenue = np.mean(valid_revenue)
median_revenue = np.median(valid_revenue)
std_dev_revenue = np.std(valid_revenue)
print(f"Mean Revenue is: {mean_revenue:.2f}")
print(f"Median Revenue is: {median_revenue:.2f}")
print(f"Standard Deviation of Revenue is: {std_dev_revenue:.2f}")
```

```python
customer_satisfaction_array = df["Customer_Satisfaction"].values
print(f"Customer Satisfaction Array: {customer_satisfaction_array}")
```

```python
valid_cs = customer_satisfaction_array[~np.isnan(customer_satisfaction_array) & (customer_satisfaction_array>=0)]
print(f"Valid Customer Satisfaction Array: {valid_cs}")
```

```python
cs_mean = np.mean(valid_cs)
cs_median = np.median(valid_cs)
cs_stddev = np.std(valid_cs)
print(f"Customer Satisfaction Mean: {cs_mean:.2f}")
print(f"Customer Satisfaction Median: {cs_median:.2f}")
print(f"Standard Deviation of Customer Satisfaction: {cs_stddev:.2f}")
```

```python
increased_revenue = valid_revenue * 1.1
print(f"Increased Revenue Array:\n\n {increased_revenue}")
```

```python
high_revenue = valid_revenue[valid_revenue>7000]
print("High Revenue Values (>7000): \n\n", high_revenue)
```

```python
high_revenue_post_increase = increased_revenue[increased_revenue>7000]
print("High Revenue After 10% Increase (>7000)\n\n", high_revenue_post_increase)
```
